사용자의 이름, 나이, 성별을 출력하는 함수 print_info를 작성하세요. 이 때, 모든 매개변수는 키워드 매개변수로만 받도록 하세요.

```python
# 키워드 매개변수
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')

print_info(name='oz', age=20, gender='fem')
```

0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
n = 10
odd_numbers = (i for i in range(1, n+1, 2))

print(type(odd_numbers))
print(odd_numbers)
for num in odd_numbers:
    print(num)
```

```python
# 2nd answer # 참고
def odd_numbers(n):
    for i in range(n):
        if i % 2 != 0:
            yield i

print(odd_numbers(10))
print(next(odd_numbers(10)))
print(next(odd_numbers(10)))
print(next(odd_numbers(10)))
print(next(odd_numbers(10)))
print(next(odd_numbers(10)))
print()
for num in odd_numbers(10):
    print(num)
```

```python
# 3rd answer
def odd_numbers(n):
    return (i for i in range(1, n+1, 2))

print(type(odd_numbers(10)))
print(odd_numbers(10))
for num in odd_numbers(10):
    print(num)
```

사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max(numbers):
    print('최소값:', min(numbers), '최댓값:', max(numbers))

num_tuple = tuple(map(int, input("3개의 숫자를 입력하세요: ").split(' ')))
min_max(num_tuple)
```

```python
# 참고: 가변매개변수 이용
def min_max(*numbers):
    print(type(numbers))  # <class 'tuple'>
    # numbers_tuple = tuple(numbers)  # unnecessary. parameter is already a tuple
    print("최소값:", min(numbers), "최대값:", max(numbers))

min_max(10, 4, 9)
```

> Ch. 29

주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(number):
    if number % 2 != 0:
        return False  # early return
    return True

print(is_even(1))
print(is_even(2))
```

> Ch. 30

팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(str_list):
    for str in str_list:
        print(len(str))

print_lengths(['ppel', 'banana', 'dragonfruit'])
```

> Ch. 32

두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1, str2):
    return str1 if len(str1) >= len(str2) else str2

print(longer_string("apple", "banana"))
```
