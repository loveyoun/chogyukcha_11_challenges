사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요.

[예외처리]
- 입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError
- 위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드를 작성

````python
try:
    x, y = map(int, input('두 개의 숫자: ').split(' '))
    z = x / y
except ValueError:
    print('숫자를 입력하세요.')
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
else:
    print(f'{x} / {y} = {z}')
````

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
numbers = []
while True:
    try:
        num_list = input("숫자를 하나 입력하세요 ('done'이면 종료): ")
        if num_list == 'done':
            print(numbers)
            break
        numbers.append(int(num_list))
    except ValueError:
        print(f"잘못된 값: {num_list} 숫자를 입력해주세요.")
````

사용자 정의 예외 NegativeNumberError를 생성하고, 사용자로부터 숫자를 입력받아 음수일 경우 NegativeNumberError를 발생시키는 프로그램을 작성하세요.

- 예외 처리 구문을 사용하여 음수 입력 시 "양수를 입력해주세요."라는 메시지를 출력

````python
class NegativeNumberError(Exception):
    def __init__(self, message) -> None:
        super().__init__(message)

try:
    number = int(input("양수를 입력하세요: "))
    if number < 0:
        raise NegativeNumberError("양수를 입력해주세요.")
except NegativeNumberError as e:
    print('!예외 발생!', e)
````

````python
# 참고 코드
class NegativeNumberError(Exception):
    def __init__(self, message) -> None:
        self.name = "NegativeNumberError"
        self.message = message
        super().__init__(message)

    def __str__(self) -> str:
        return f"{self.name} : {self.message}"

try:
    number = int(input("양수를 입력하세요: "))
    if number < 0:
        raise NegativeNumberError("양수를 입력해주세요.")
except NegativeNumberError as e:
    print(e)
````

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def elapsed(func):
    def wrapper():  # *args, **kwargs
        start_time = time.time()
        result = func()
        end_time = time.time()
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
    return wrapper

@elapsed
def dummy():
    sum = 0
    for i in range(1000000):
        sum += i
    return sum

dummy()
````
