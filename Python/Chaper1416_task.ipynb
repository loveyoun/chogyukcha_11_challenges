 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]

result = []
for i in range(len(arr1)):
    tmp = []
    for j in range(len(arr1[i])):
        tmp.append(arr1[i][j] + arr2[i][j])
    result.append(tmp)


""" result = [[c1 + c2 for c1, c2 in zip(r1, r2)] for r1, r2 in zip(arr1, arr2)] """

print(result)
```

__[문제5]__

양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 합니다.   
예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.   
주어진 x를 이용해 x가 하샤드 수인지 아닌지 검사하는 코드를 작성해주세요

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
x = 17
xsum = 0
for i in range(len(str(x))):
    xsum += int(str(x)[i])
if x % xsum == 0:
    print("true")
else:
    print("false")


# num_list = list(map(int, str(x)))
# result = x % sum(num_list) == 0
# print(result)
```
