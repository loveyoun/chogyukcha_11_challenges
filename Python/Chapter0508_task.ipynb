__[문제 1: 첫 번째 사냥]__

```python
characters = {
    'magician' : {
        'strength' : 10,
        'dexterity' : 20,
        'intelligence' : 80,
        'health' : 50
    },
    'warrior' : {
        'strength' : 80,
        'dexterity' : 30,
        'intelligence' : 20,
        'health' : 100
    },
    'thief' : {
        'strength' : 40,
        'dexterity' : 80,
        'intelligence' : 30,
        'health' : 60
    },
    'healer' : {
        'strength' : 20,
        'dexterity' : 40,
        'intelligence' : 70,
        'health' : 70
    }
}

print(characters['magician']['intelligence'] >= 50)
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
XPs = {
    'magician' : 120,
    'warrior' : 150,
    'thief' : 110,
    'healer' : 90
}

for i in XPs:  # == XPs.keys()
    print(f'{i}\'s been upgraded?   {XPs[i] >= 100}')
```
