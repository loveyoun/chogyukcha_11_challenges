```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    "",
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ã„±
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ã„±
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ã„±
     |   /ã…£
     |
    ---"""
]


class HangmanGame:

    global words, hangman_pics

    def __init__(self):
        print('===== HANGMAN game just started... =====')

    def initGame(self):
        self.idx = random.randint(0,4)
        self.word = words[self.idx]
        self.size = len(self.word)
        self.uline = ['_'] * self.size
        self.chance = 7

        print(f'\nANS: {self.word}\n')

    def display(self, flag):
        print(hangman_pics[7 - self.chance])  # 7,6,5,4,3,2,1,0 = 0(default),1,2,3,4,5,6,7(-1)
        if flag == 0 or flag == 2:  # ì‹¤íŒ¨ flag = 0, í¬ê¸° = 2
            print(f"The answer was... '{''.join(self.word)}'")
        else:  # ìŠ¹ë¦¬ flag = 1, defalt = -1
            print(' '.join(self.uline))

    def play(self):
        rs = True
        while rs:
            # ê°’ ì´ˆê¸°í™” #
            self.initGame()
            flag, turn = -1, 0

            while self.chance > 0:
                isWrong = True
                self.display(flag)  # ê·¸ë¦¼,ë‹¨ì–´ ë§¤ë²ˆ ì¶œë ¥
                n = input('> Guess the letter(q to quit): ')

                # validation #
                if n == 'q':
                    flag = 2
                    break
                if n in self.uline:
                    print('@ WARNING: ì´ë¯¸ ë§ì¶˜ ë‹¨ì–´ì…ë‹ˆë‹¤. ë‹¤ë¥¸ ë‹¨ì–´ë¥¼ ì…ë ¥í•˜ì„¸ìš”.\n')
                    continue

                turn += 1
                for i in range(self.size):
                    if self.word[i] == n:
                       self.uline[i] = n
                       isWrong = False

                # ì •ë‹µ ì—¬ë¶€ #
                if isWrong:
                    self.chance -= 1  # 7ë²ˆì§¸ í‹€ë¦¬ë©´ chance = 0
                    if self.chance > 0:
                        print(f'[í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ê¸°íšŒëŠ”... {self.chance-1}ë²ˆ]\n')
                else:
                    print("[ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”.]\n")

                # ìŠ¹ë¦¬ ì—¬ë¶€ #
                if '_' not in self.uline:
                    flag = 1  # ìŠ¹ë¦¬
                    break


            flag = 0 if flag == -1 else flag  # ì‹¤íŒ¨ flag = 0
            self.display(flag)
            if flag == 1:
                print(f'ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤! ì´ {turn}ë²ˆ ì‹œë„í•˜ì…¨ë„¤ìš”.\n')
            elif flag == 0:
                print('you HANGEDDD the MAN:(\n')
            else:  # í¬ê¸° flag = 2
                print('you GAVE UP....LoL')
                break

            while True:
                try:
                    rs = input('> Press y to continue, or q to quit: ')
                    if rs == 'q':
                        rs = False
                        break
                    elif rs == 'y':
                        break
                    else:  # @ rs!='y'(True) or rs!='q'
                        raise Exception
                except Exception:
                    print('ì˜³ì€ ì…ë ¥ì„ í•´ì£¼ì„¸ìš”.')


        print('=========================================')


if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
