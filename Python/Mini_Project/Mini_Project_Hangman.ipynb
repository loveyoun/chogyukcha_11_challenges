#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

10번만에 맞히지 못하면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
hangman_pics = [
    "",
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |   /ㅣ
     |
    ---"""
]


class HangmanGame:

    global words, hangman_pics

    def __init__(self):
        print('===== HANGMAN game just started... =====')

    def initGame(self):
        self.idx = random.randint(0,4)
        self.word = words[self.idx]
        self.size = len(self.word)
        self.uline = ['_'] * self.size
        self.chance = 7

        print(f'\nANS: {self.word}\n')

    def display(self, flag):
        print(hangman_pics[7 - self.chance])  # 7,6,5,4,3,2,1,0 = 0(default),1,2,3,4,5,6,7(-1)
        if flag == 0 or flag == 2:  # 실패 flag = 0, 포기 = 2
            print(f"The answer was... '{''.join(self.word)}'")
        else:  # 승리 flag = 1, defalt = -1
            print(' '.join(self.uline))

    def play(self):
        rs = True
        while rs:
            # 값 초기화 #
            self.initGame()
            flag, turn = -1, 0

            while self.chance > 0:
                isWrong = True
                self.display(flag)  # 그림,단어 매번 출력
                n = input('> Guess the letter(q to quit): ')

                # validation #
                if n == 'q':
                    flag = 2
                    break
                if n in self.uline:
                    print('@ WARNING: 이미 맞춘 단어입니다. 다른 단어를 입력하세요.\n')
                    continue

                turn += 1
                for i in range(self.size):
                    if self.word[i] == n:
                       self.uline[i] = n
                       isWrong = False

                # 정답 여부 #
                if isWrong:
                    self.chance -= 1  # 7번째 틀리면 chance = 0
                    if self.chance > 0:
                        print(f'[틀렸습니다. 남은 기회는... {self.chance-1}번]\n')
                else:
                    print("[단어를 맞추셨군요. 더 힘내봐요.]\n")

                # 승리 여부 #
                if '_' not in self.uline:
                    flag = 1  # 승리
                    break


            flag = 0 if flag == -1 else flag  # 실패 flag = 0
            self.display(flag)
            if flag == 1:
                print(f'축하합니다! 단어를 맞추셨습니다! 총 {turn}번 시도하셨네요.\n')
            elif flag == 0:
                print('you HANGEDDD the MAN:(\n')
            else:  # 포기 flag = 2
                print('you GAVE UP....LoL')
                break

            while True:
                try:
                    rs = input('> Press y to continue, or q to quit: ')
                    if rs == 'q':
                        rs = False
                        break
                    elif rs == 'y':
                        break
                    else:  # @ rs!='y'(True) or rs!='q'
                        raise Exception
                except Exception:
                    print('옳은 입력을 해주세요.')


        print('=========================================')


if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```

```python
""" REFACTORED-예시 코드 """

import random

words = ["apple", "banana", "orange", "grape", "lemon"]
hangman_pics = [
    "",
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|ㄱ
     |   /ㅣ
     |
    ---"""
]


class HangmanGame:
    # 클래스가 만들어 질 때 자동으로 호출되는 메소드
    # (생성자 메소드라고 부릅니다)
    def __init__(self):
        self.word = random.choice(words)  # 랜덤으로 단어 선택
        self.guesses = set()  # 추측된 글자 저장
        self.attempts = 7  # 시도 횟수 = 6

    def display(self):
        # 단어의 현재 상태 표시
        result = ""
        for char in self.word:
            if char in self.guesses:
                result += char + " "
            else:
                result += "_ "

        return result

    def play(self):
        # 게임을 시작합니다.
        # 시도 횟수가 0이 될 때까지 반복합니다.
        while self.attempts > 0:
            print(hangman_pics[7 - self.attempts])
            print(self.display())

            guess = input("글자를 추측해보세요: ").lower()

            if guess in self.guesses:  # 이미 맞춘 단어
                print("이미 추측한 글자입니다.")
            elif guess in self.word:  # 새로운 단어(정답)
                self.guesses.add(guess)

                # 승리
                if set(self.word) == self.guesses:  # set(self.word).issubset(self.guesses):
                    print(f"축하합니다! 단어를 맞추셨습니다: {self.word}")
                    break

                print("단어를 맞추셨군요. 더 힘내봐요")
            else:  # 새로운 단어(오답)
                self.attempts -= 1

                # 실패
                if self.attempts == 0:
                    print(f"게임 오버. 정답은: {self.word}")
                else:
                    print(f"틀렸습니다. 남은 시도 횟수: {self.attempts-1}")


if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
